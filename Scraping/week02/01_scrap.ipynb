```python
from urllib.request import urlopen
import numpy as np
import pandas as pd
import json
```

```python
url = "https://crm.nissan.co.th/mapi/evar/evmarkers.aspx"
```

```python
response = urlopen(url)
```

```python
data_json = json.loads(response.read())
```

```python
df = pd.json_normalize(data_json, record_path =['data'])
df
```

```python
df.columns
```

```python
df.info()
```

```python
df2 = df[['networkname','stationname','addeng','plgtypename','lat','lng']]
df2.rename(columns = {'networkname':'Operator', 'stationname':'Station', 'addeng':'Address', 'plgtypename':'Charger', 'lat':'Lattitude', 'lng':'Longitude'}, inplace = True)
```

```python
df2.info()
```

```python
df2.head()
```

```python
df3 = df2[df2['Operator'] == "NISSAN EV"]
df3
```

```python
df3['Province'] = df3['Address'].apply(lambda x: x.split(" ")[-2])
```

```python
df3
```

```python
df3.to_csv('NISSAN_EV_STATION.csv',encoding="utf-8-sig", index=False)
```
